# How has the average salary (in USD) evolved yearly from 2020 to 2024? 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data
df = pd.read_csv("data_science_salaries.csv")
# Group by work_year and calculate the average salary
average_salary_per_year = df.groupby("work_year")["salary_in_usd"].mean()
# Plotting the trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_salary_per_year.index, y=average_salary_per_year.values, marker='o', color='b')
plt.title("Average Salary Trend (2020–2024)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by job_title and calculate the average salary
average_salary_by_job = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False)
# Get the top 5 highest and lowest-paying job titles
top_10_highest = average_salary_by_job.head(10)
top_10_lowest = average_salary_by_job.tail(10)
# Plot the highest-paying job titles
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_highest.values, y=top_10_highest.index, palette="Blues_r")
plt.title("Top 10 Highest-Paying Job Titles", fontsize=16)
plt.xlabel("Average Salary (USD)", fontsize=10)
plt.ylabel("Job Titles", fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
# Plot the lowest-paying job titles
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_lowest.values, y=top_10_lowest.index, palette="Reds_r")
plt.title("Top 10 Lowest-Paying Job Titles", fontsize=16)
plt.xlabel("Average Salary (USD)", fontsize=10)
plt.ylabel("Job Titles", fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Which job titles offer the highest and lowest salaries? (for india )

```python
# Filter the data for India
df_india = df[df['company_location'] == 'India']
# Group by job_title and calculate the average salary in USD
average_salary_by_job_india = df_india.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False)
# Get the top 5 highest and lowest-paying job titles in India
top_10_highest_india = average_salary_by_job_india.head(10)
top_10_lowest_india = average_salary_by_job_india.tail(10)
# Plot the highest-paying job titles in India
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_highest_india.values, y=top_10_highest_india.index, palette="Blues_r")
plt.title("Top 10 Highest-Paying Job Titles in India", fontsize=16)
plt.xlabel("Average Salary (USD)", fontsize=10)
plt.ylabel("Job Titles", fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
# Plot the lowest-paying job titles in India
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_lowest_india.values, y=top_10_lowest_india.index, palette="Reds_r")
plt.title("Top 10 Lowest-Paying Job Titles in India", fontsize=16)
plt.xlabel("Average Salary (USD)", fontsize=10)
plt.ylabel("Job Titles", fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# How do salaries vary by experience level (junior, mid-level, senior)?

```python
# Group by experience level and calculate the average salary
average_salary_by_experience = df.groupby("experience_level")["salary_in_usd"].mean().sort_values(ascending=False)
# Plotting the salary distribution by experience level
plt.figure(figsize=(8, 6))
sns.barplot(x=average_salary_by_experience.index, y=average_salary_by_experience.values, palette="viridis")
plt.title("Average Salary by Experience Level", fontsize=16)
plt.xlabel("Experience Level", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# What are the highest-paying company locations? 

```python
# Group by company_location and calculate the average salary
average_salary_by_location = df.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False)
# Get the top 7 highest-paying company locations
top_7_highest_location = average_salary_by_location.head(7)
# Plotting the highest-paying company locations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_7_highest_location.values, y=top_7_highest_location.index, palette="Blues_r")
plt.title("Top 7 Highest-Paying Company Locations", fontsize=16)
plt.xlabel("Average Salary (USD)", fontsize=12)
plt.ylabel("Company Locations", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

# Which work model (remote, hybrid, on-site) offers the highest salaries?

```python
# Group by work_models and calculate the average salary
average_salary_by_work_model = df.groupby("work_models")["salary_in_usd"].mean().sort_values(ascending=False)
# Plotting the average salary for each work model
plt.figure(figsize=(8, 6))
sns.barplot(x=average_salary_by_work_model.index, y=average_salary_by_work_model.values, palette="viridis")
plt.title("Average Salary by Work Model", fontsize=16)
plt.xlabel("Work Model", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Does employment type (full-time, contract, freelance) affect salary significantly? 

```python
# Group by employment_type and calculate the average salary
average_salary_by_employment_type = df.groupby("employment_type")["salary_in_usd"].mean().sort_values(ascending=False)
# Plotting the salary distribution by employment type
plt.figure(figsize=(8, 6))
sns.barplot(x=average_salary_by_employment_type.index, y=average_salary_by_employment_type.values, palette="coolwarm")
plt.title("Average Salary by Employment Type", fontsize=16)
plt.xlabel("Employment Type", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# How does company size influence pay for specific job titles or work models? 

```python
# Group by company_size and job_title, then calculate the average salary
average_salary_by_company_size_job = df.groupby(["company_size", "job_title"])["salary_in_usd"].mean().unstack()
# Plotting the salary distribution for different job titles across company sizes
average_salary_by_company_size_job.plot(kind="bar", figsize=(12, 8), colormap="viridis")
plt.title("Average Salary by Company Size and Job Title", fontsize=16)
plt.xlabel("Company Size", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Job Titles", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Do larger companies pay better than smaller ones?

```python
# Group by company_size and calculate the average salary
average_salary_by_company_size = df.groupby("company_size")["salary_in_usd"].mean().sort_values(ascending=False)
# Plotting the salary distribution by company size
plt.figure(figsize=(8, 6))
sns.barplot(x=average_salary_by_company_size.index, y=average_salary_by_company_size.values, palette="magma")
plt.title("Average Salary by Company Size", fontsize=16)
plt.xlabel("Company Size", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
